## 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.


```python
def plot_ellipse(a=2, b=3, x0=0, y0=0, points=200, color='black'):
    x = np.linspace(0, a, points)
    y = np.sqrt(a**2 - x**2)*b/a
    x = np.concatenate((x[::-1],-x,-x[::-1],x))
    y = np.concatenate((y[::-1],y,-y[::-1],-y))
    if x0:
        x = x + x0
    if y0:
        y = y + y0
    plt.plot(x,y, color=color)


# окружность - частный случай элипса, где a==b==R
def plot_circle(R=1, x0=0, y0=0, points=200, color='black'):
    plot_ellipse(a=R, b=R, x0=x0, y0=y0, points=points, color=color)

    
def plot_hyperbole(a=3, b=5, x0=0, y0=0, points=200, lim=1, color='black'):
    x = np.linspace(a, a+lim, points)
    y = np.sqrt(x**2 - a**2)*b/a
    x1 = np.concatenate(( x[::-1], x))
    y1 = np.concatenate(( y[::-1],-y))
    x2 = np.concatenate((-x[::-1],-x))
    y2 = np.concatenate(( y[::-1],-y))
    if x0:
        x1 = x1 + x0
        x2 = x2 + x0
    if y0:
        y1 = y1 + y0
        y2 = y2 + y0 
    plt.plot(x1,y1,color=color)
    plt.plot(x2,y2,color=color)
    
    
plt.figure(figsize=(6,6))
plt.xlim(-15,10)
plt.ylim(-15,10)
plt.grid(False)

plot_ellipse(a=1,b=3,color='r')
plot_ellipse(a=7, b=10)
plot_circle(R=2, x0=3.25,y0=3.25)
plot_circle(R=2, x0=-3.25,y0=3.25)
plot_circle(R=0.8, x0=3,y0=4,color='b')
plot_circle(R=0.8, x0=-3,y0=4,color='b')
for _x0 in np.linspace(-4.5,4.5,20):
    plot_circle(R=0.3, x0=_x0, y0=-5, points=5, color='gold')
plot_ellipse(a=5,b=0.5, y0=-5,  color='r')
plot_hyperbole(a=7.6,b=5,lim=0.7)
plot_hyperbole(a=7,b=3,lim=1)
plot_hyperbole(a=3.5,b=3,x0=0,y0=-10, lim=1.9)
for _a in np.linspace(0.15,3.5,15):
    plot_hyperbole(a=_a,b=1,x0=0,y0=8.9, lim=0.25,color='brown')
for _a in np.linspace(0.25,4.5,15):
    plot_hyperbole(a=_a,b=0.45,x0=0,y0=-3.5, lim=0.2,color='brown')
plt.show()
```

Нарисуйте график функции: `y(x) = k∙cos(x – a) + b` для некоторых (2-3 различных) значений параметров k, a, b


```python
def plot_func(a=0, b=0, k=1, nT=1, color=None, label=''):
    x = np.arange(-np.pi*nT, np.pi*nT,0.1)
    y = k * np.cos(x - a) + b
    if label=='':
        label = f'a={a}, b={b}, k={k}'
    plt.plot(x,y,color=color,label=label)
    
    
fig = plt.figure(figsize=(6,6))
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.grid(True)

plot_func(color='black', nT=1.5, label='default')
plot_func(1,1,3,2)
plot_func(0,0,-0.5,2)
plot_func(-1,-3,-2,2)
plt.legend()
plt.show()
```

1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def dec_to_polar(x,y, radian=True):
    r = np.sqrt(x*x+y*y)
    phi = np.arcsin(y/r)
    return (phi,r) if radian else (phi*180/np.pi, r)
```

```python
phi, r = dec_to_polar(1,1)
print('r=',r)
print('phi=',phi,'радиан')
```

```python
phi, r = dec_to_polar(1,1,radian=False)
print('r=',r)
print('phi=',round(phi),'градусов')
```

```python
phi, r = dec_to_polar(0,1,radian=False)
print('r=',r)
print('phi=',phi,'градусов')
```

#### 1) Решите систему уравнений:
* y = x² – 1
* exp(x) + x∙(1 – y) = 1

##### 1 способ

```python
x = np.arange(-2,5,0.01)
x = x[x.nonzero()]
    
plt.ylim(-2,20)
plt.xlim(-2,5)
plt.grid(True)

plt.plot(x,(x**2)-1,color='g')
plt.plot(x,(np.exp(x) + x - 1)/x, color='b')
plt.scatter(0, 2.0, color='black')

# проверка ответов
plt.scatter(2.61814557342307, 5.854686243819049, color='red')
plt.scatter(4.200105839767363, 16.640889065285183, color='red')
plt.scatter(-1.5818353528959237, 1.5022030836713227, color='red')

plt.show()
```

```python
from scipy.optimize import fsolve
def eq(p):
    x, y = p
    return ((x**2)-1-y,np.exp(x)+x*(1-y)-1)

x1,y1 = fsolve(eq,(3,-2))
print(x1,y1)

x2,y2 = fsolve(eq,(6,-2))
print(x2,y2)

x3,y3 = fsolve(eq,(-2,-1))
print(x3,y3)

#особый случай, пояснения ниже:
x4,y4 = fsolve(eq,(0,3))
print(x4,y4)
```

#### Ответ:
В точке `x=0` функция `y = (exp(x) + x – 1) / x` не определена, поэтому уравнению `exp(x) + x∙(1 – y) = 1` при `x=0` удовлетворяют любые `y`, а для системы уравнений удовлетворяют соответственно только `y = x² – 1 = 0² – 1 = -1` 
* x1 = `2.61814557342307`    y1 = `5.854686243819049`
* x2 = `4.200105839767363`   y2 =`16.640889065285183`
* x3 = `-1.5818353528959237` y3 =`1.5022030836713227`
* x4 = `0` y4 =`-1`

#### 2 способ
* Подстановка `y` из первого уравнения во второе, 
* поиск корней `x` получившегося уравнения, 
* подстановка полученнных корней в первое уравнения для получения корней `y`.

```python
def y(x):
    return np.exp(x) + 2*x - x**3 - 1
    

x = np.arange(-2,5,0.01)

plt.ylim(-4,4)
plt.xlim(-2,5)
plt.grid(True)

plt.plot(x,y(x), color='b')

# проверка ответов
plt.scatter(2.61814557, 0, color='red')
plt.scatter(4.20010584, 0, color='red')
plt.scatter(-1.58183535, 0, color='red')
plt.scatter(0, 0, color='red')

plt.show()
```

```python
xs = [fsolve(y,-1), fsolve(y,2.5), fsolve(y,4.3),fsolve(y,0)]
ys = [(_x**2)-1 for _x in xs]

for i, _x, _y in zip([1,2,3,4], xs, ys):
    print(f'x{i}={_x}, y{i}={_y},')
```

#### 2) Решите систему уравнений и неравенств:
* y = x² – 1
* exp(x) + x∙(1 – y) > 1

```python
def y(x):
    return np.exp(x) + 2*x - x**3 - 1
    

x = np.arange(-3,5,0.01)

plt.ylim(-4,25)
plt.xlim(-3,5)
plt.grid(True)
plt.hlines(0,-3,5,color='r')
plt.plot(x,(x**2)-1,color='g', label='x**2')
plt.plot(x,y(x), color='b',label='exp(x) + 2*x - x**3 - 1')
plt.legend()

x1 = np.linspace(-3,-1.5818,10)
plt.plot(x1,(x1**2)-1,color='black', linewidth=5 )

x2 = np.linspace(0,2.62,10)
plt.plot(x2,(x2**2)-1,color='black', linewidth=5 )

x3 = np.linspace(4.20,5,10)
plt.plot(x3,(x3**2)-1,color='black', linewidth=5 )

plt.vlines((-1.58, 0, 2.618, 4.20), ymin=-10, ymax=25, linewidth=1, color='r', linestyles='dashed' )

plt.show()
```

Неравенству в системе удовлетовряют все х, при которых функция `y = exp(x) + 2∙x - x**3 - 1` больше нуля.

Корни уравнения `exp(x) + 2∙x - x**3 - 1 = 0` возьмем из предыдущего расчета. 

* x1 = `2.61814557342307` 
* x2 = `4.200105839767363`
* x3 = `-1.5818353528959237` 
* x4 = `0` 

Анализируя график и полученные граничные корни, получим решение системы:

`y = x**2 -1` при: `x: (-inf,-1.58) или (0, 2.62) или (4.20, +inf)`
