### Задача 1

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя уровень значимости 0.025 и правостороннюю критическую область, проверить данную гипотезу при условии, что в выборке из 100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.



```python
a = 17  # нулевая гипотеза
a_mean = 17.5  # средний радиус в выборке
alpha = 0.025  # уровень значимости
n = 100  # размер выборки
var = 4  # дисперсия
T = (a_mean-a) / np.sqrt(var/n)  # значение статистики
```

Правосторонней критической области $(t_{(1-\alpha)}, +\infty)$ соответствует правый бесконечный хвост стандартного нормального распределения, отсеченный критической точкой со значением $2\sigma$ (так как двум сигмам соответствует 95% вероятности, $\sigma=1$.

```python
t = 2
```

Если результат $Т$ при проведении статистического теста превысит значение $t_{(1-\alpha)}$, то нулевая гипотеза будет отвергнута.

```python
print('Результат теста, Т =', T)
print(f'Критическая область: ({t}, +inf)')
print('Нулевая гипотеза ПРИНИМАЕТСЯ') if T > t else print('Нулевая гипотеза ОТВЕРГАЕТСЯ')
```

### Задача 3

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:

`202, 203, 199, 197, 195, 201, 200, 204, 194, 190`

Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что уровень значимости равен 1%?

```python
samples = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
alpha = 0.01
a = 200

a_mean = samples.mean()  # 198.5
n = samples.size  # 10
std = samples.std(ddof=1)  # 4.453463071962462
t = t_dist.ppf(1-alpha/2, df=n-1)  # 3.2498355440153697
T = (a_mean - a) / (std/np.sqrt(n))  # -1.0651074037450896

print('Результат теста, Т =', T.round(2))
print(f'Критическая область: (-inf, {-t.round(2)}) U ({t.round(2)}, +inf)')
print('Утверждение продавца ПРИНИМАЕТСЯ') if (-t < T < t) else print('Утверждение продавца ОТВЕРГАЕТСЯ')
```
